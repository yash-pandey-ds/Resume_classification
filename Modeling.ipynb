```python
import numpy as np
import pandas as pd
import pickle as pk
```

```python
df = pd.read_csv(r'cleaned_resume_data.csv')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(df.Cleaned_Resume)
required_text = tfidf.transform(df.Cleaned_Resume)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(required_text,df.Type,test_size=0.2,random_state=124)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print('The accuracy of the model using KNN is:',accuracy_score(y_test,y_pred)*100)
```

```python
pk.dump(model,open('final_model.sav','wb'))
```

```python
n_neighbors = [1,2,3,4,5]   #a
weights = ['uniform', 'distance'] #b
leaf_size = range(1,41)  #c
metric = [ 'manhattan', 'euclidean', 'chebyshev']  #d
algorithm = ['auto', 'ball_tree','kd_tree','brute']   #e
p = [1,2]   #f
```

```python
# knn_model = KNeighborsClassifier(n_neighbors=)
accuracy_list = []
for a in n_neighbors:
    for b in weights:
        for c in leaf_size:
            for d in metric:
                for e in algorithm:
                    for f in p:
                        knn_model = KNeighborsClassifier(n_neighbors=a,weights=b,leaf_size=c,algorithm=e,p=f)
                        knn_model.fit(X_train, y_train)
                        y_pred = knn_model.predict(X_test)
                        accuracy = accuracy_score(y_test, y_pred) * 100
                        if accuracy not in accuracy_list:
                            accuracy_list.append(accuracy)
                        print(f'n_neighbors: {a}\nweights: {b}\nleaf_size: {c}\nmetric: {d}\nalgorithm: {e}\np: {f}\naccuracy: {accuracy}')
                        print()
```

```python
accuracy_list
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

modelLR = LogisticRegression()
modelLR.fit(X_train,y_train)
```

```python
y_pred_LR = modelLR.predict(X_test)
accuracy_score(y_test,y_pred_LR)*100
```

```python
penalty = ['l1', 'l2', 'elasticnet']    #a
C = [0.0001, 0.001, 0.01, 0.1, 1]       #b
solver = ['liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']      #c
```

```python
import itertools
accuracy_list2 = []
for pen, c_val, sol in itertools.product(penalty, C, solver):
    try:
        model = LogisticRegression(penalty=pen, C=c_val, solver=sol, max_iter=1000)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)*100
        if accuracy not in accuracy_list2:
            accuracy_list2.append(accuracy)
        print(f"Penalty: {pen},\nC: {c_val},\nSolver: {sol},\nAccuracy: {accuracy:.4f}")
        print()
    except Exception as e:
        print('Failed')
        print()
```
